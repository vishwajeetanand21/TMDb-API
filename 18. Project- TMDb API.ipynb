```python
for i in myTagline:
    if(i==''):
        print('Empty')
    else:
        print(i)
```

# TMDb 3.4

```python
# Finding the total number of pages
response=requests.get('https://api.themoviedb.org/3/tv/airing_today',params={'api_key':myAPIKey,'language':'en'})

print(response.url)

pythonData=json.loads(response.text)

totalPages=pythonData['total_pages']
print(totalPages)

# We have to print the names of all the TV shows, and there are total 18 pages in total
```

```python
# Printing the names of all TV shows from all the pages

for a in range(1,totalPages+1):

    response=requests.get('https://api.themoviedb.org/3/tv/airing_today',params={'api_key':myAPIKey,'language':'en','page':a})

# print(response.url)

    pythonData=json.loads(response.text)

    result=pythonData['results']

    for i in range(len(result)):
        if(result[i]['original_language']=='en'):
            print(result[i]['name'])
```
